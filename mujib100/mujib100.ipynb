```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
# creating soup
def create_soup(url):
    response = requests.get(url)
    soup     = BeautifulSoup(response.content, 'html.parser')
    return soup

url = 'https://mujib100.gov.bd/pages/mujib/photo-archive_bn.html'
soup = create_soup(url)
```

```python
# list of all class having photo & text
img_element_list = soup.find_all(class_ = 'col-sm-4 mb-3 wow zoomIn')
```

```python
# function for text processing

def text_processing(element):
    list_1 = element.p.text.strip().split('\n')
    list_2 = [i.strip() for i in list_1]
    full_text = (' ').join(list_2)
    return full_text

    # full code in one line
    # (' ').join([i.strip() for i in element.p.text.strip().split('\n')])
```

```python
# above function will work like this
(' ').join([i.strip() for i in img_element_list[34].p.text.strip().split('\n')])
```

```python
# image url of first element
# this is not full url as it contains ./../../
img_element_list[0].a['href']
```

```python
# Creating full url of image
t1 = 'https://mujib100.gov.bd/'
t2 = img_element_list[0].a['href'][8:]
img_url = t1 + t2
img_url
```

```python
# list to store image url and text
all_data_list = []

# iterating every class to get image url and text
for element in img_element_list:
    dic = {}

    t1 = 'https://mujib100.gov.bd/'
    t2 = element.a['href'][8:]
    img_url = t1 + t2
    img_url

    dic['url']  = img_url
    dic['text'] = text_processing(element)

    all_data_list.append(dic)


all_data_list
```

```python
!pip install pymupdf
import fitz
# fitz is included in pymupdf

#pymupdf documentation
#https://pymupdf.readthedocs.io/en/latest/index.html
#https://documentation.help/PyMuPDF/
```

```python
all_data_list[0]['url']
```

```python
# creating pdf file object
# later we will use below line
# page = doc.new_page()
# to insert page

doc = fitz.open()
# this doc object has no page at this moment

for item in all_data_list:
    link = item['url']
    response = requests.get(link)

    # save response content as .jpg file
    with open('image.jpg', 'wb') as f:
        f.write(response.content)

    # inserting new page
    page = doc.new_page()

    # where to put image 
    # we use Rect (rectengular) 
    # Rect(x0,y0, x1,y1)
    rect = fitz.Rect(40, 50, 560, 520)     
    #The first two numbers are regarded as the “top left” corner (x0,y0) 
    #and (x1,y1) as the “bottom right” one


    page.insert_image(rect, filename = 'image.jpg')

    doc.save("mujib100images.pdf")
os.remove('image.jpg')
```
